```
import pandas as pd
import numpy as np
from statsmodels.api import OLS
from statsmodels.tools.tools import add_constant
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```
data = pd.read_csv("/content/drive/My Drive/Data/us.csv") 
```

```
confirmed = data.cases
```

```
x = np.arange(len(confirmed))
x = add_constant(x)
#creating a corresponding timeline (162 days)
```

```
model = OLS(np.log(confirmed), x) # Ordinary Least Square regression using historical data starting from January 21, 2020
```

```
result = model.fit()
```

```
result.summary()
```

```
plt.plot(
    np.exp(result.predict(x)),
    label="Fitted exp. function"      
)
plt.plot(confirmed, ".", label="Reported cases, US")
plt.legend()
plt.xlabel("Days from Januray 21, 2020")
plt.ylabel("Number of Cases")
plt.show()
#plotting graph showing comparison of fitted function with real-life data
```

```
Exponential_Predictions_162 = np.exp(result.predict(x[:162]))
len(Exponential_Predictions_162)
```
